```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img=cv2.imread("/content/drive/MyDrive/Summer 2026/108503-V1/Age-related-Macular-Degeneration-Retinal-OCT-images/OCTID_AMD/AMRD1.jpeg",0)

#NlMean
dst = cv2.fastNlMeansDenoising(img)

#Gaussian
blur = cv2.GaussianBlur(img,(5,5),1000)


plt.figure(tight_layout=True)

plt.subplot(221)
plt.imshow(img,cmap="gray")
plt.title("original")

plt.subplot(222)
plt.imshow(dst,cmap="gray")
plt.title("Non-local Mean")

plt.subplot(223)
plt.imshow(blur,cmap="gray")
plt.title("Gaussian")
plt.show()
```

```python
plt.figure(tight_layout=True)

plt.subplot(221)
plt.imshow(img,cmap="gray")
plt.title("original")

equ = cv2.equalizeHist(img)

plt.subplot(222)
plt.imshow(equ,cmap="gray")
plt.title("HE")

# HE after Non-local mean
equN = cv2.equalizeHist(dst)

plt.subplot(223)
plt.imshow(equN,cmap="gray")
plt.title("HE after Non-local")


# Histogram Equalization after Gaussian Blur
equG = cv2.equalizeHist(blur)

plt.subplot(224)
plt.imshow(equG,cmap="gray")
plt.title("HE after Gaussian Blur")

plt.show()
```

```python
#CLAHE
clipl=2.0 #clip limit
tgs=(8,8) #tile grid size

plt.figure(tight_layout=True)

plt.subplot(221)
plt.imshow(img, cmap="gray")
plt.title("original")

#CLAHE
clahe = cv2.createCLAHE(clipLimit=clipl,tileGridSize =tgs)
cl1 = clahe.apply(img)

plt.subplot(222)
plt.imshow(cl1, cmap="gray")
plt.title("CLAHE")

#CLAHE after Non-local
clahe = cv2.createCLAHE(clipLimit=clipl,tileGridSize =tgs)
cl2 = clahe.apply(dst)

plt.subplot(223)
plt.imshow(cl2, cmap="gray")
plt.title("CLAHE after Non-local")


#CLAHE after Gaussian
clahe = cv2.createCLAHE(clipLimit=clipl,tileGridSize =tgs)
cl3 = clahe.apply(blur)

plt.subplot(224)
plt.imshow(cl3, cmap="gray")
plt.title("CLAHE after Gaussian")

plt.show()



```

```python
#NlMean
dst = cv2.fastNlMeansDenoising(img)

#Gaussian
blur = cv2.GaussianBlur(img,(5,5),1000)

#Median
med = cv2.medianBlur(img,5)

plt.figure(tight_layout=True)

plt.subplot(221)
plt.imshow(img,cmap="gray")
plt.title("original")

plt.subplot(222)
plt.imshow(dst,cmap="gray")
plt.title("Non-local Mean")

plt.subplot(223)
plt.imshow(blur,cmap="gray")
plt.title("Gaussian")

plt.subplot(224)
plt.imshow(med, cmap="gray")
plt.title("Median")

plt.show()
```

```python
#CLAHE after Gaussian
clahe = cv2.createCLAHE(clipLimit=clipl,tileGridSize =tgs)
cl4 = clahe.apply(med)

plt.imshow(cl4, cmap="gray")
plt.title("CLAHE after Median")

plt.show()
```
