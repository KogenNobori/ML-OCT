# OCTDL

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img=cv2.imread("/content/drive/MyDrive/Summer 2026/OCTDL Optical Coherence Tomography Dataset for Image-Based Deep Learning Methods/OCTDL/DME/dme_1057060_1.jpg",0)



plt.figure(tight_layout=True)

#Original
plt.subplot(221)
plt.imshow(img,cmap="gray")
plt.title("Original")

#Gaussian
Gaus = cv2.GaussianBlur(img,(5,5),10) #How should I choose the parameters?

plt.subplot(222)
plt.imshow(Gaus,cmap="gray")
plt.title("Gaussian blurring")

#CLAHE after Gaussian
clipl = 2.0
tgs = (8,8)

clahe = cv2.createCLAHE(clipLimit=clipl,tileGridSize =tgs)
clage_Gaus_DL = clahe.apply(Gaus)

plt.subplot(223)
plt.imshow(clage_Gaus_DL, cmap="gray")
plt.title("CLAHE after Gaussian blurring")


#Edge detection with Canny method

edge_clage_Gaus_DL= cv2.Canny(clage_Gaus_DL, 40, 200)

plt.subplot(224)
plt.imshow(edge_clage_Gaus_DL, cmap="gray")
plt.title("Edges after Gaussian and CLAHE")


plt.show()
```
